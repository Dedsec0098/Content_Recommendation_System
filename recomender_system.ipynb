```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj) :
        L.append(i['name'])
    return L
# Output of movies.iloc[0].genre will be a string and hence feeding it to this function as obj will throw us type errors 
# So to convert it from str to list we will use ast library of python called ast 
```

```python
movies['genres'] = movies['genres'].apply(convert)  # we will apply convert function to all the genres where for every movie we get a list of genres of that movie
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)  # we will apply same function to keywords also and get it cleaned
```

```python
movies.head() # Now we need to the same thing on cast but we need only first 3 names because they are the most imp ones
```

```python
movies.cast[0]  # this is the rae form of cast and we need to extract from here
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj) :
        if counter != 3 :
            L.append(i['name'])
            counter+=1
        else :
            break
    return L              # this function will give us starting 3 names only from cast
```

```python
movies['cast'].apply(convert3)
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
movies['crew'][0]  # Now we will extract the director's name from the crew and remove everything else
```

```python
def fetch_director(obj) :
    L = []
    for i in ast.literal_eval(obj) :
        if i['job'] == 'Director' :
            L.append(i['name'])
            break 
    return L                    # we will get all the directors of the different movies
```

```python
movies['crew'].apply(fetch_director)
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]  # we can see that overview is given as a string and we need to convert it to a list
```

```python
movies['overview'] = movies['overview'].fillna('').apply(lambda x: x.split())  # this will split each word of a string to form a lit 
# and it will fill NAN values with an empty string ' '
```

```python
movies.head()
```

```python
# DATA TRANSFORMATION - we will remove space in between words because lets say there is 'Sam Worthington' and 'Sam Mendes' and if spaces not
# removed then the model will think all the Sam are similar and may remmend different than what we are looking henve we will apply transformation
```

```python
movies['genre'] = movies['genres'].apply(lambda x : [i.replace(" " ,"") for i in x])  # this will replace " " with ""
movies['keywords'] = movies['keywords'].apply(lambda x : [i.replace(" " ,"") for i in x])
movies['cast'] = movies['cast'].apply(lambda x : [i.replace(" " ,"") for i in x])
movies['crew'] = movies['crew'].apply(lambda x : [i.replace(" " ,"") for i in x])
```

```python
movies.head()
```

```python
movies['tag'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']  # we will concatinate all of it
```

```python
movies.head()
```

```python
new_df = movies[['movie_id' , 'title' , 'tag']]  # we will form a new dataframe which will have only movie_id , title , tag
```

```python
new_df.head()
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x : ' '.join(x))
```

```python
new_df['tag'][0]
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x : x.lower())  # applied lower function to all the values of tag
```

```python
new_df.head()
```

```python
# Text Verctorization : In this step we will convert all the tags of each movie into verctor(cordinates) and then we will plot in a 2d space
# and to find related movies we will find all the cordinate points that nearby the orginal ones 
```

```python
# We will use technique called bag of words to convert text into vectors by usign the following steps : 
# 1) We will combine all the tags of movies and find 5000 common words that are there in tags 
# 2) We  will then select tag of each movies and then we will compare it against every common words to find how many times is has occured
# 3) Then we will form a vector of all the 5000 movies with all the occurace of 5000 common words and its shape would be ( 5000 x 5000 )

# This will convert all the tags of the movie to a vector which can plotted in 2D space 
# Our aim is to reduce the dimensionality of the matrix, means that instead of 5000 words we can use 2k or 1k common words to map which -
#  - will give higher accuracy
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')  # counts most common 5k words and remove stop words ( and , if , I etc)
```

```python
vectors = cv.fit_transform(new_df['tag']).toarray()  # it will store all the common words
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
# we have many words like action , actions , actinable etc and all of the are simmilar in meaning hence we will reduce it using NLTk
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()  # object of class PorterStemmer
```

```python
def stem(text):   # helper function of class 
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
ps.stem('loved')
```

```python
new_df['tag'] = new_df['tag'].apply(stem)
```

```python
cv.get_feature_names_out()
```

```python
# Now we will be calculating between each vectors points on 2d space and to calculate distance between points 
# Eucladian distance is not a good measure to find distance between two points when we have high dimensionality points
# Hence we will be using Cosine distance which is angle between two vectors point 
```

```python
# We will use cosine similarity function in scikit learn library to calculate cosine distance
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)  # we will apply cosine similarity function to all the vectors calculated
```

```python
similarity.shape  # this is a 4809 x 4809 matrix which will be having similarity of all the movies with each movie and all the diagonal will
# be 1 of this matrix because every movie will be similar to itself hence similarity will be 1 for it.
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]  # we will get the index of the movie from new_df
    distances  = similarity[movie_index]  # then get the cosine similarity vector for that particular indexed movie
    movie_list = sorted(list(enumerate(distances)), reverse = True , key = lambda x:x[1])[1:6]

    for i in movie_list :
        print(new_df.iloc[i[0]].title)
     
```

```python
# we will now sort the distances in descending order to get most simillar movies but the problem is when we sort we will loose the index 
# values of those movies and hence we will use enumarate function that keeps counter
```

```python
# sorted(list(enumerate(similarity[0])), reverse = True , key = lambda x:x[1])[1:6]  # this function will do sorting with enumerate function
# and it will apply it on the values not the indexes using lambda function and extract top 5 using [1:6]
```

```python
recommend('Unstoppable')
```
